based on the data above, row 0-6 and 39-54 is description about the data. So, we must clean the data.
Note: row of description the data not always in row 0-6 and 39-54. Can be different, dependent a many days per month

```python
# concate all data become 1 data
path = 'D:\\DIAJAR\\magang\\belajar program\\Python\\Projek\\Data Iklim\\data\\' # directory file into a folder years
data_list = [] # will input data frame to data_list

# because there is 37 years from 1984 - 2020
for i in range(37):
    year = 1984 + i      # define year         
    new_path = path + '{}\\'.format(str(year))  # new path to read excel folder
    # because there is 12 file excel per folder
    for i in range(12):
        # there is 2 case name file
        if i == 0:
            data_i = pd.read_excel(new_path + 'laporan_iklim_harian.xlsx')
            data_i = data_i.drop(labels=[0,1,2,3,4,5,6])  # drop row 0-6, it consistent for all data
            data_i = data_i.dropna(thresh=2)              # drop lower row section 
            data_i = [data_i]                             # input data to list
            data_list += data_i                           # list of data
        else:
            data_i = pd.read_excel(new_path + 'laporan_iklim_harian ({}).xlsx'.format(str(i)))
            data_i = data_i.drop(labels=[0,1,2,3,4,5,6])
            data_i = data_i.dropna(thresh=2)
            data_i = [data_i]
            data_list += data_i 
new_data = pd.concat(data_list, axis=0)                   # concate all data
```

```python
# rename column
new_data.columns = ['Tanggal', 'Tn', 'Tx', 'Tavg', 'RH_avg', 'RR', 'ss', 'ff_x', 'ddd_x', 'ff_avg', 'ddd_car']
```
